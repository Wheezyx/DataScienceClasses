```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("Funkcja liniowa, kwadratowa, trzeciego rzÄ™du")
x = np.linspace(0, 2, 50)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

```python
print("Cosinus")
x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
cars = pd.read_csv('samochody1tys.csv')
```

```python
print('Pogrupowanie wzglÄ™dem marki, zliczenie oraz utworzenie wykresu sÅ‚upkowego')
plot = cars.groupby('marka')['marka'].count().sort_values().plot.bar(figsize=(15, 15))
plot.set_ylabel('Ilosc')
```

```python
print('Grupowanie po roku produkcji, zliczenie i wykreÅ›lenie wykresy koÅ‚owego.')
cars.groupby('rok_produkcji')['rok_produkcji'].count().plot.pie(figsize=(15, 15), legend=True, autopct='%1.1f%%')
```
