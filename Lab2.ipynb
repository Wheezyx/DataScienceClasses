```python
print("Korzystając z wbudowanej integracji pandas z matplotliba, generujemy 1000 liczb i 1000 dat, stosujemy funckję cumsum i wykrślamy wykres")
ts = pd.Series(np.random.randn(1000),
                index=pd.date_range('1/1/2000', periods=1000))
 

ts = ts.cumsum()

ts.plot()
```

```python
print("Dodajemy dodatkowe kolumny i więcej danych, tworzymy wykres i wykreślamy na nim dane")
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index,
                      columns=['A', 'B', 'C', 'D'])
   

df = df.cumsum()

plt.figure()
df.plot()
plt.legend(loc='best')
```
